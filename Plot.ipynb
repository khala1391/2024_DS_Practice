```python
group = ("aa", "bb", "cc", "dd")
value = (1,2,3,4)
plt.axis("equal")
plt.pie(value, 
        labels=group,
        startangle=90, 
        autopct="%1.2f%%", 
        explode=(0,0,0,0.4)
       )

plt.show();
```

```python
gr = np.array(["aa", "bb", "cc", "dd"])
value = (1,2,3,4)
plt.axis("equal")

expl = np.zeros(gr.size)
expl[2] = 0.2
plt.pie(value, 
        labels=gr,
        startangle=90, 
        autopct="%1.2f%%", 
        explode= expl
       )

plt.show();
```

```python
gr = np.array(["aa", "bb", "cc", "dd"])
value = (1,2,3,4)
plt.axis("equal")

expl = np.zeros(gr.size)
# expl[2] = 0.2
expl[np.where(gr == "bb")] = 0.2

plt.pie(value, 
        labels=gr,
        startangle=90, 
        autopct="%1.2f%%", 
        explode=expl
       )

plt.show();
```

```python
gr = np.array(["aa", "bb", "cc", "dd"])
value = (1,2,3,4)
plt.axis("equal")

expl = np.zeros(gr.size)
# expl[2] = 0.2
expl[np.where(gr == "bb")] = 0.2
col = ["red", "green", "pink", "orange"]
plt.pie(value, 
        labels=gr,
        startangle=90, 
        autopct="%1.2f%%", 
        explode=expl,
        colors =col,
       )

plt.show();
```

```python
from scipy import stats
```

```python
x = [170, 165, 180, 155, 163, 160, 175]
y = [80, 65, 75, 47, 50, 52, 74]
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
print(slope, intercept, r_value**2, p_value)
l =slope*min(x)+ intercept , slope*max(x)+ intercept

plt.scatter(x,y)
plt.plot([min(x),max(x)],l, linestyle='--', color='green', alpha =0.5)
plt.xlabel("height(cm)")
plt.ylabel("weight(kg)")

# t = "y = {:.2f} + {:.2f}x, r^2 = {:.4f}, p = {:.4f}".format(intercept, slope, r_value**2, p_value)
# t = r"$y = {:.2f} + {:.2f}x, r^2 = {:.4f}, p = {:.4f}$".format(intercept, slope, r_value**2, p_value)
t = r"${} = {:.2f} + {:.2f}x, r^2 = {:.4f}, p = {:.4f}$".format("\^{y}",intercept, slope, r_value**2, p_value)
plt.title(t)
plt.show()
```

# normpdf

https://github.com/materialsproject/pymatgen/issues/1657

from matplotlib.mlab import normpdf -> from scipy.stats import norm <br>
normpdf -> norm.pdf

```python
import matplotlib.pyplot as plt
# import matplotlib.mlab as mlab
import numpy as np
from scipy.stats import norm
```

normpdf -> norm.pdf

```python
plt.figure(figsize =(3,2))
mu, sigma = 0, 1
x = np.linspace(-4,4,100)
plt.plot(x,norm.pdf(x,mu,sigma))
plt.show()
```

```python
plt.figure(figsize =(3,2))
# mu, sigma = 0, 1
mu = [0,0,0,2]
sigma = [0.5,1,2,1]

x = np.linspace(-4,4,100)
for  m,s in zip(mu, sigma):
    plt.plot(x,norm.pdf(x,m,s))

# plt.plot(x,norm.pdf(x,mu,sigma))
plt.show()
```

```python
plt.figure(figsize =(6,4))
# mu, sigma = 0, 1
mu = [0,0,0,2]
sigma = [0.5,1,2,1]

x = np.linspace(-4,4,100)
for  m,s in zip(mu, sigma):
    plt.plot(x,norm.pdf(x,m,s),
            label = r"$\mu={:.1f}, \sigma={:.1f}$".format(m,s))
# plt.plot(x,norm.pdf(x,mu,sigma))

plt.legend()
plt.title("normal distribution curve")

plt.show()
```

```python
plt.figure(figsize =(6,4))
# mu, sigma = 0, 1
mu = [0,0,0,2]
sigma = [0.5,1,2,1]
linestyle = [ "-","--","-.",":"]

x = np.linspace(-4,4,100)
for  m,s,ls in zip(mu, sigma, linestyle):
    plt.plot(x,norm.pdf(x,m,s),
            label = r"$\mu={:.1f}, \sigma={:.1f}$".format(m,s),
            linestyle=ls)

# plt.plot(x,norm.pdf(x,mu,sigma))

plt.legend()
plt.title("probability density function")
plt.axvline(0, color="0.8", linestyle="--")
plt.axvline(2, color="0.5", linestyle="--")

plt.show()
```

# subplot2grid

```python
fig = plt.figure()
ax1 = plt.subplot2grid((2, 6), (0, 0), rowspan=2, colspan=4)
ax2 = plt.subplot2grid((2, 6), (0, 4), rowspan=1, colspan=2)
ax3 = plt.subplot2grid((2, 6), (1, 4), rowspan=1, colspan=2)
x = np.arange(6)
ax1.plot(x, np.random.randint(1, 11, x.size))
ax2.bar(x, np.random.randint(1, 11, x.size), color="orange")
ax3.scatter(x, np.random.randint(1, 11, x.size), color="red")
ax1.set_title("line")
ax2.set_title("bar")
ax3.set_title("scatter")
fig.tight_layout()
plt.show()
```
